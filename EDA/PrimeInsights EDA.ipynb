###8. Define Redundant Methods -

````python
# Creating a function to explode multi value columns-
def explode_columns(df, columns, reset_index=True):
    """
    Explodes one or multiple list-like columns in a DataFrame into separate rows.

    Parameters:
    - df (pd.DataFrame): The input DataFrame
    - columns (str or list of str): Column name(s) to explode
    - reset_index (bool): Whether to reset the index after exploding (default True)

    Returns:
    - pd.DataFrame: DataFrame with specified columns exploded
    """
    # Make a copy to avoid modifying the original DataFrame
    df_copy = df.copy()

    # Ensure columns is a list for uniform handling
    if isinstance(columns, str):
        columns = [columns]

    # Check if all specified columns exist in DataFrame
    for col in columns:
        if col not in df_copy.columns:
            raise ValueError(f"Column '{col}' not found in DataFrame.")

    # Explode columns one by one
    for col in columns:
        df_copy = df_copy.explode(col)

    # Optionally reset the index for clean output
    if reset_index:
        df_copy.reset_index(drop=True, inplace=True)

    return df_copy
````

````python
def get_value_counts(df, column_name, top_n=None):
    """
    Returns a DataFrame containing the value counts of a categorical column.

    Parameters:
    - df (pd.DataFrame): Input dataframe
    - column_name (str): Column name whose value counts are needed
    - top_n (int or None): If provided, returns only the top_n values

    Returns:
    - pd.DataFrame: A DataFrame with unique values and their counts
    """
    value_counts = df[column_name].value_counts().reset_index()
    value_counts.columns = [column_name.title(), 'Count']

    if top_n:
        return value_counts.head(top_n)
    return value_counts
````

````python
def group_count_unstack(df, index_col, column_col, fill_value=0, observed=True):
    """
    Returns a pivot-style DataFrame using groupby + size + unstack pattern.
    Useful for counting occurrences of combinations of two categorical variables.

    Parameters:
    - df (pd.DataFrame): The input dataframe
    - index_col (str): Column to be used as index (rows)
    - column_col (str): Column to be used as columns (will be unstacked)
    - fill_value (int): Value to fill for missing combinations (default is 0)
    - observed (bool): Whether to include only observed combinations (default True)

    Returns:
    - pd.DataFrame: A pivot-style DataFrame with count values
    """
    return df.groupby([index_col, column_col], observed=observed).size().unstack(fill_value=fill_value)
````

````python
def group_agg_unstack(df, index_cols, value_col, agg_func='median', fill_value=None):
    """
    Groups the DataFrame by one or more index columns, aggregates a value column,
    and unstacks the result if two grouping columns are provided.

    Parameters:
    - df (DataFrame): The input DataFrame.
    - index_cols (str or list): One or two columns to group by.
    - value_col (str): The column to aggregate.
    - agg_func (str or callable): Aggregation function ('mean', 'median', 'sum', etc.).
    - fill_value (any, optional): Value used to fill missing values after unstacking.

    Returns:
    - DataFrame: Grouped (and unstacked if applicable) result.
    """
    # Ensure index_cols is a list even if a single string is passed
    if isinstance(index_cols, str):
        index_cols = [index_cols]

    # Perform groupby aggregation with observed=True to avoid FutureWarning
    grouped = df.groupby(index_cols, observed=True)[value_col].agg(agg_func)

    # If two grouping columns, unstack the last one for better readability
    if len(index_cols) == 2:
        return grouped.unstack(fill_value=fill_value)
    else:
        return grouped
````

````python
def top_n_list(df, column, n=5):
    """
    Returns the first N entries from a specified column in a DataFrame as a list.

    Parameters:
    - df (pd.DataFrame): Input DataFrame.
    - column (str): Column name from which to extract values.
    - n (int): Number of top values to return (default is 5).

    Returns:
    - list: First N values from the specified column as a list.

    Note:
    - The function does not sort or clean NaNs.
    - Assumes the DataFrame is already ordered appropriately if needed.
    """
    # Check if column exists
    if column not in df.columns:
        raise ValueError(f"Column '{column}' not found in DataFrame.")

    # Return the first n values as a list (including any NaNs)
    return df[column].head(n).tolist()
````

````python
def filter_column_value(df, column, value):
    """
    Filters a DataFrame based on a specific column's value(s).

    Parameters:
    - df (pd.DataFrame): The input DataFrame.
    - column (str): The name of the column to filter on.
    - value (str, list, or None): The value or list of values to filter for.
        - If None: returns rows where the column is null.
        - If str: returns rows where the column exactly matches the string.
        - If list or other iterable: returns rows where column value is in the list.

    Returns:
    - pd.DataFrame: A filtered DataFrame based on the condition.
    """

    # Case 1: Filter where value is None (NaN)
    if value is None:
        return df[df[column].isnull()]

    # Case 2: Filter exact match for a string
    elif isinstance(value, str):
        return df[df[column] == value]

    # Case 3: Filter using list or iterable of values
    else:
        return df[df[column].isin(value)]
````

````python
def club_rows(df, column_name, filter_list, club_name, agg_func, agg_column, groupby_columns=None):
    """
    Groups a DataFrame by one or more columns, replacing values not in `filter_list`
    in `column_name` with `club_name`, then aggregates using the provided `agg_func`.

    Parameters:
    - df (pd.DataFrame): Input DataFrame.
    - column_name (str): The column to filter and group values on.
    - filter_list (list): List of values to keep; others are grouped into `club_name`.
    - club_name (str): Name for the grouped "others" category.
    - groupby_columns (str, list, or None): Column(s) to group by in addition to `column_name`.
    - agg_func (str or callable): Aggregation method like 'sum', 'mean', or a custom function.
    - agg_column (str): The column on which to perform aggregation.

    Returns:
    - pd.DataFrame: Aggregated and grouped DataFrame.
    """
    import pandas as pd

    # Input validation
    if not isinstance(df, pd.DataFrame):
        raise ValueError("df must be a pandas DataFrame.")
    if column_name not in df.columns:
        raise ValueError(f"'{column_name}' not found in DataFrame.")
    if agg_column not in df.columns:
        raise ValueError(f"'{agg_column}' not found in DataFrame.")
    if not callable(agg_func) and not isinstance(agg_func, str):
        raise ValueError("agg_func must be a string (like 'sum') or a callable function.")

    # Normalize groupby_columns to a list
    if groupby_columns is None:
        groupby_columns = []
    elif isinstance(groupby_columns, str):
        groupby_columns = [groupby_columns]
    elif not isinstance(groupby_columns, list):
        raise ValueError("groupby_columns must be a string, list of column names, or None.")

    # Check that all groupby_columns exist in the DataFrame
    for col in groupby_columns:
        if col not in df.columns:
            raise ValueError(f"'{col}' not found in DataFrame.")

    # Replace non-top items with club_name
    df = df.copy()
    df['__grouped__'] = df[column_name].where(df[column_name].isin(filter_list), club_name)

    # Group and aggregate
    group_cols = groupby_columns + ['__grouped__']

    try:
        result = df.groupby(group_cols, observed=False)[agg_column].agg(agg_func).reset_index()
        result = result.rename(columns={'__grouped__': column_name})
        return result
    except Exception as e:
        raise RuntimeError(f"Failed to group and aggregate: {e}")
````

This comprehensive analysis of Amazon Prime Video’s content library offers valuable insights into genre trends, regional contributions, platform performance metrics, and viewer preferences. It is evident that the platform’s content ecosystem is primarily driven by a few dominant genres—Drama, Comedy, Action, Thriller, and Romance—with Drama leading consistently across countries and formats. However, a significant portion of content still falls under the broadly classified “Others” category, highlighting an opportunity to enhance metadata quality and content classification for improved discoverability and recommendation accuracy.

In terms of content types, movies considerably outnumber shows, although shows are gradually gaining ground, particularly post-2015. This signals a potential growth area for Amazon Prime to diversify its episodic content portfolio, especially by leveraging emerging genres and regional storytelling formats. Runtime patterns, too, reflect regional storytelling nuances—with Indian movies tending to be longer and more elaborate, while countries like the US and Japan favour shorter, more concise formats.

A country-wise breakdown indicates that the United States continues to be the primary content contributor, but countries like India, the United Kingdom, France, and Japan are making notable strides, especially in genre-specific content. Interestingly, Japan and Korea stand out for their high TMDB popularity despite moderate IMDb scores, indicating the influence of loyal niche audiences and younger demographics.

Analysis of IMDb and TMDB scores reveals a divergence in viewer sentiment across platforms. While IMDb scores tend to be consistent but moderate, TMDB popularity is often driven by virality or recency, particularly for genres like Thriller. This underlines the importance of adopting platform-specific engagement strategies and avoiding assumptions based on aggregate rating systems.

Temporal trends from 2000 to 2022 reveal that Amazon Prime has aggressively expanded its movie catalogue, with a sharp increase in content from 2016 onwards, peaking in 2021. Meanwhile, the growth of TV shows—although slower—has been consistent and appears to respond to shifting audience preferences and production capacities. Genre-specific growth patterns also indicate that Comedy, Romance, and Thriller have seen substantial spikes in recent years.

Lastly, correlations between features like IMDb Votes, TMDB Popularity, and Release Year reaffirm that newer content tends to be more widely engaged with, even when critical ratings are average. This points towards a data-backed need to distinguish between viewer engagement metrics and viewer satisfaction scores, and to align marketing, acquisition, and recommendation strategies accordingly.

In conclusion, the analysis not only sheds light on the current state of content diversity and consumption on Amazon Prime but also offers actionable insights for strategic planning, including improving genre taxonomy, region-specific content investment, targeted marketing, and data-driven retention models.
